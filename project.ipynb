# Get the data

```python
df2018 = pd.read_csv('2018-QS-World-University-Rankings.csv', encoding='ISO-8859-1', header = [0,1])
df2019 = pd.read_csv('2019-QS-World-University-Rankings.csv', encoding='ISO-8859-1', header = [0,1])
df2020 = pd.read_csv('2020-QS-World-University-Rankings.csv', encoding='ISO-8859-1', header = [0,1])
```

# Resolving Columns

```python
def resolve_columns(df: pd.core.frame.DataFrame) -> list:
    cols = []
    rankNumber = 1
    for column in df:
        if column[0].startswith('Unnamed'):
            if column[1].startswith('RANK'):
                cols.append(column[1] + '_' + str(rankNumber))
                rankNumber +=1
            else:
                cols.append(column[1])
        elif column[1].startswith('Unnamed'):
            if column[0].startswith('RANK'):
                cols.append(column[0] + '_' + str(rankNumber))
                rankNumber +=1
            else:
                cols.append(column[0])
        else:
            cols.append(column[0] + '_' + column[1])
    return cols
```

```python
df2018.columns = resolve_columns(df2018)
df2019.columns = resolve_columns(df2019)
df2020.columns = resolve_columns(df2020)
```
